* Importing required modules, downloading and reading the raw dataset

```python
import pandas as pd
url = "https://www.ers.usda.gov/webdocs/DataFiles/80591/DataDownload2015.xlsx?v=0"
df = pd.read_excel(url, sheet_name='Food Access Research Atlas')
df.sample(5)
```

* Subsetting the raw dataset to filter out all the states but New York state

```python
NYdf = df.loc[df.State == "New York"]
NYdf.sample(5)
```

* Selecting necessary columns (i.e. CensusTract and LAhalfand10)
* Filtering out all the boroughs but the five boroughs in New York City

```python
SelectedColumns = ["CensusTract", "LAhalfand10"]
FiveBoroughs = ["Bronx", "Kings", "New York", "Queens", "Richmond"]
FoodAccess = NYdf[NYdf["County"].isin(FiveBoroughs)].loc[:,SelectedColumns].reset_index(drop=True)
FoodAccess.sample(5)
```

* Saving results to a new file

```python
FoodAccess.to_csv("FoodAccess.csv")
```
